```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats

%matplotlib inline
```

```python
df = pd.read_csv('athlete_events.csv')
```

```python
df2 = pd.read_csv('noc_regions.csv')
```

```python
df = df.merge(df2, on = 'NOC')
```

```python
new_df = df.dropna(subset = ['Age', 'Height', 'Weight'])
new_df
```

```python
all_women = new_df[new_df['Sex'] == 'F'].copy()
```

```python
all_women['Weight(lb)'] = all_women['Weight'].map(lambda x: x*2.2)
```

```python
all_women['Height(in)'] = all_women['Height'].map(lambda x: x/2.54)
all_women['Height(in)'] = all_women['Height(in)'].round(1)
```

```python
all_women['Height(m)'] = all_women['Height'].map(lambda x: x/100)
```

```python
all_women['BMI'] = (all_women['Weight']/(all_women['Height(m)']**2)).round(1)
```

```python
all_women
```

```python
# Women participated in the 1900 Games in Paris, France. 
# Twenty-two women (2.2 per cent) out of a total of 997 athletes competed in five sports. 
# These sports were tennis, sailing, croquet, equestrian and golf.
# First gold medal won by a women was in 1920
# 1996 Olympics was considered the Summer of the Womens due to the US Women's dominance in sports 
```

```python
# Question 1: Is there a difference in BMI's of women from 1996 ('Summer of Women') and 2016?
# Null hypothesis: there is no difference in BMI's from women participants in 1996 and 2016
# Alternative hypothesis: there is a difference in BMI's from women participants in 1996 and 2016
# Test-statistic: Z-test
# Alpha: .05
```

```python
all_women_1996 = all_women[all_women['Year'] == 1996]
```

```python
all_women_2016 = all_women[all_women['Year'] == 2016]
```

```python
pop_mean_1996_bmi = all_women_1996['BMI'].mean()
pop_std_1996_bmi = all_women_1996['BMI'].std()
```

```python
# Taking sample of year 2016 for comparison
sample_of_all_women_2016_bmi = all_women_2016['BMI'].sample(n=30)

sample_of_all_women_2016_mean_bmi = np.mean(sample_of_all_women_2016_bmi)

sample_of_all_women_2016_se_bmi = pop_std_1996_bmi/np.sqrt(len(sample_of_all_women_2016_bmi))
```

```python
pop_mean_1996_bmi
```

```python
sample_of_all_women_2016_mean_bmi
```

```python
z1 = (sample_of_all_women_2016_mean_bmi - pop_mean_1996_bmi)/sample_of_all_women_2016_se_bmi
```

```python
p1 = 1 - stats.norm.sf(z1)
```

```python
# Conclusion: pvalue = 0.89, so we fail to reject the null hypothesis. 
# There is a no difference in the BMI's of women from 1996 ('Summer of Women') and 2016
```

```python
#Observe distribution of BMIs of 1996 vs 2016
sns.set()
fig = plt.figure(figsize = (15, 10))
sns.distplot(all_women_1996['BMI'], color = 'blue', label = '1996 women')
sns.distplot(all_women_2016['BMI'], color = 'green', label = '2016 women')
plt.title("BMI of Females: 1996 vs 2016", fontsize = 20)
plt.xlabel("BMI of Women", fontsize = 20)
plt.legend(prop={'size': 20})
plt.savefig("BMI_females_1996_vs_2016") 
```

```python
# Question 2: Is there a difference in age of women from 1996 ('Summer of Women') and 2016?
# Null hypothesis: there is no difference in age from women participants in 1996 and 2016
# Alternative hypothesis: there is a difference in age from women participants in 1996 and 2016
# Test-statistic: Z-test
# Alpha: .05
```

```python
pop_mean_1996_age = all_women_1996['Age'].mean()
pop_std_1996_age = all_women_1996['Age'].std()
```

```python
# Taking sample of year 2016 for comparison
sample_of_all_women_2016_age = all_women_2016['Age'].sample(n=30)

sample_of_all_women_2016_mean_age = np.mean(sample_of_all_women_2016_age)

sample_of_all_women_2016_se_age = pop_std_1996_age/np.sqrt(len(sample_of_all_women_2016_age))
```

```python
pop_mean_1996_age
```

```python
sample_of_all_women_2016_mean_age
```

```python
z2 = (sample_of_all_women_2016_mean_age - pop_mean_1996_age)/sample_of_all_women_2016_se_age
z2
```

```python
p2 = 1 - stats.norm.sf(z2)
p2
```

```python
# Conclusion: pvalue = 0.99, so we fail to reject the null hypothesis. 
# There is a no difference in the ages of women from 1996 ('Summer of Women') and 2016
```

```python
#Observe distribution of BMIs of 1996 vs 2016
sns.set()
fig = plt.figure(figsize = (15, 10))
sns.distplot(all_women_1996['Age'], color = 'blue', label = 'Age during 1996', bins = 30)
sns.distplot(all_women_2016['Age'], color = 'green', label = 'Age during 2016')
plt.title("Age of Females: 1996 vs 2016", fontsize = 20)
plt.xlabel("Age of Women", fontsize = 20)
plt.legend(prop={'size': 20})
plt.savefig("Ages_females_1996_vs_2016")
```
